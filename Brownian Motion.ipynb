```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n = 1000; T = 1; dt = T/n
W = np.zeros(n+1)
t = np.linspace(0,T,n+1)
t
```

```python
for i in range(1,n+1):
    W[i] = W[i-1] + np.sqrt(dt)*np.random.randn(1)
W
```

```python
plt.plot(t,W)
plt.xlabel('time'); plt.ylabel('W'); plt.title('Brownian Path')
plt.show()
```
